## Exploratory Data Analysis (EDA)

```python
# Check for missing values
print("Missing Values:")
print(df.isnull().sum())
print(f"\nTotal missing values: {df.isnull().sum().sum()}")
print(f"Percentage of missing values: {(df.isnull().sum().sum() / (df.shape[0] * df.shape[1]) * 100):.2f}%")

# Identify the target column (last column)
target_col = df.columns[-1]
print(f"\nTarget Column: {target_col}")

# Separate features and target
X = df.iloc[:, :-1]
y = df.iloc[:, -1]

print(f"\nFeatures shape: {X.shape}")
print(f"Target shape: {y.shape}")

# Analyze target variable
print("\n")
print("TARGET VARIABLE ANALYSIS")
print(f"\nTarget variable: {target_col}")
print(f"Unique values: {y.nunique()}")
print(f"\nClass distribution:")
print(y.value_counts().sort_index())
print(f"\nClass distribution (percentages):")
print(y.value_counts(normalize=True).sort_index() * 100)
```

Visualizing the class distribution helps identify severe class imbalance

```python
counts = y.value_counts().sort_index()
perc = (y.value_counts(normalize=True).sort_index() * 100)

df = pd.DataFrame({
    "Severity Class": counts.index.astype(str),
    "Count": counts.values,
    "Percentage": perc.values
})

palette = px.colors.qualitative.Set2

# --- Subplots ---
fig = make_subplots(
    rows=1, cols=2,
    subplot_titles=("Class Distribution (Count)", "Class Distribution (Percentage)"),
    horizontal_spacing=0.08
)

fig.add_trace(
    go.Bar(
        x=df["Severity Class"],
        y=df["Count"],
        marker=dict(color=palette[0]),
        hovertemplate="<b>%{x}</b><br>Count: %{y:,}<extra></extra>"
    ),
    row=1, col=1
)

fig.add_trace(
    go.Bar(
        x=df["Severity Class"],
        y=df["Percentage"],
        marker=dict(color=palette[3] if len(palette) > 3 else palette[1]),
        hovertemplate="<b>%{x}</b><br>Percentage: %{y:.2f}%<extra></extra>"
    ),
    row=1, col=2
)


fig.update_layout(
    template="plotly_white",
    height=450,
    width=1100,
    title=dict(text="Class Distribution", x=0.5),
    bargap=0.35,
    margin=dict(l=60, r=30, t=70, b=140),
    font=dict(size=12),
    showlegend=False
)


fig.update_xaxes(tickangle=40, tickfont=dict(size=11))
fig.update_yaxes(title_text="Count", row=1, col=1)
fig.update_yaxes(title_text="Percentage (%)", row=1, col=2)


fig.update_xaxes(categoryorder="array", categoryarray=df["Severity Class"])

fig.show()
```

```python
# Analyze feature types and distributions
print("FEATURE ANALYSIS")

# Identify numeric and categorical features
numeric_features = X.select_dtypes(include=[np.number]).columns.tolist()
categorical_features = X.select_dtypes(include=['object']).columns.tolist()

print(f"\nNumeric features ({len(numeric_features)}): {numeric_features}")
print(f"\nCategorical features ({len(categorical_features)}): {categorical_features}")

# Check for constant or near-constant features
print("\n" + "-"*80)
print("Feature Variance Analysis:")
for col in numeric_features:
    unique_ratio = X[col].nunique() / len(X)
    print(f"{col}: {X[col].nunique()} unique values ({unique_ratio*100:.2f}% of data)")
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import math

if len(numeric_features) > 0:
    n_cols = 4
    n_rows = math.ceil(len(numeric_features) / n_cols)

    fig = make_subplots(
        rows=n_rows,
        cols=n_cols,
        subplot_titles=numeric_features,
        horizontal_spacing=0.06,
        vertical_spacing=0.10
    )

    for i, col in enumerate(numeric_features):
        row = i // n_cols + 1
        col_idx = i % n_cols + 1

        fig.add_trace(
            go.Histogram(
                x=X[col],
                nbinsx=30,
                marker=dict(
                    color="rgba(99, 110, 250, 0.75)",
                    line=dict(color="rgba(0,0,0,0.4)", width=1)
                ),
                hovertemplate="Value: %{x}<br>Count: %{y}<extra></extra>"
            ),
            row=row,
            col=col_idx
        )


    total_plots = n_rows * n_cols
    for j in range(len(numeric_features), total_plots):
        row = j // n_cols + 1
        col_idx = j % n_cols + 1
        fig.update_xaxes(visible=False, row=row, col=col_idx)
        fig.update_yaxes(visible=False, row=row, col=col_idx)

    fig.update_layout(
        title=dict(
            text="Distribution of Numeric Features",
            x=0.5,
            font=dict(size=18)
        ),
        template="plotly_white",
        height=n_rows * 260,
        showlegend=False,
        bargap=0.05,
        margin=dict(t=80, l=40, r=40, b=40),
        font=dict(size=11)
    )


    fig.update_xaxes(title_text="Value")
    fig.update_yaxes(title_text="Frequency")

    fig.show()
```
